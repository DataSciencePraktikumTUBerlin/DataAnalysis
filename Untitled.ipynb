```python
import os

import numpy as np
import pandas as pd

from sklearn.decomposition import PCA

from matplotlib import pyplot as plt
import seaborn as sns

sns.set_theme(style='darkgrid')
```

```python
folder = './Big_DF/No_Notebook'

is_csv = lambda file: '.csv' in file
is_not_e_int = lambda file: 'Energy Intensity -' not in file

files = os.listdir(folder)
files = list(filter(is_csv, files))
files = list(filter(is_not_e_int, files))
```

```python
df = pd.DataFrame()

for file in files:
    file = f'{folder}/{file}'
    
    if df.size == 0:
        df = pd.read_csv(file, index_col=0)

        continue
    
    df2 = pd.read_csv(file, index_col=0)
    df2 = df2.replace('US', 'United States')
    try:
        df2 = df2.drop('variable', axis=1)
    except:
        df2
        
    df = df.merge(df2, on=['Country', 'Years'])

df = df.replace('..', np.nan).fillna(method='ffill')
df = df.rename({'Years': 'Year'}, axis=1)

df
```

```python
df.to_csv('./Big_DF/energy.csv')
```

```python
e = pd.read_csv('./Big_DF/energy.csv', index_col=0)
f = pd.read_csv('./Finance/finance.csv', index_col=0)

e = e.groupby('Year').sum()
f = f.groupby('Year').sum()

years = np.arange(2004, 2016)

e = e.loc[years]
f = f.loc[years]
```

```python
plt.figure(figsize=(12, 12))

corr = pd.concat([e, f], axis=1, keys=['e', 'f']).corr().loc['f', 'e']
# mask = np.triu(np.ones_like(corr, dtype=bool))

h = sns.heatmap(corr, vmin=-1, vmax=1, annot=True, cmap='BrBG')
h.set_xticklabels(h.get_xticklabels(), rotation=45, ha='right')
```

```python
def pca(df):
    pca = PCA(n_components=1)
    
    pca.fit(df.T)
    
    pca = pca.components_[0]
    pca = pca / pca.max()
    
    return pca
```

```python
e_pca = pca(e)
f_pca = pca(f)
```

```python
plt.figure(figsize=(16, 8))
sns.lineplot(x=f.index, y=f_pca)
sns.lineplot(x=e.index, y =e_pca)
```
