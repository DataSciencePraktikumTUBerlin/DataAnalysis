build plots for the countries we research

```python
countries = ['China', 'Germany', 'India', 'United States']

# Set figsize and loop over our countries
fig, ax = plt.pyplot.subplots(1,1, figsize=(14,6))
for country in countries:
    x = df[df['Country'] == country].Year
    y = df[df['Country'] == country]['Renewable installed PV Power (GW)']
    ax.plot(x, y)

# Set labels and legend
ax.set_title('Solar Installed Capacity', size = 20)
ax.legend(countries)
ax.xaxis.set_label_text('Years', size = 16)
ax.yaxis.set_label_text('GigaWatts', size = 16)

# Save the figure in 'IV.I_notebooks' directory. must be saved before .show()
plt.pyplot.savefig('IV.I_notebooks/Solar_Installed_capacity2.png')

# Show figure
plt.pyplot.show()
```
