build plots for the countries we research

```python
dfg = df.groupby('Year').mean()
dfg = dfg.reset_index()
```

```python

# create figure and axis objects with subplots()
fig,ax = plt.pyplot.subplots(1,1, figsize=(14,6))
ax.grid(color='lightgrey')

# make a plot
ax.plot(dfg['Year'], dfg['Electricity generation (TWh)'], color="red")
# set x-axis label
ax.set_xlabel("year",fontsize=14)
# set y-axis label
ax.set_ylabel('Electricity Generation (TWh)',color="red",fontsize=14)

# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(dfg['Year'], dfg['Total final energy consumption (TFEC) (TJ) '],color="blue")
ax2.set_ylabel('Energy Consumption (TJ)',color="blue",fontsize=14)
ax.set_title('Electricity Generation vs Energy Consumption', size = 20)

plt.pyplot.show()

# save the plot as a file
fig.savefig('IV.I_notebooks/Energy Intensity vs Wind and Solar share of electricity production.jpg', format='jpeg', dpi=100, bbox_inches='tight')
```
