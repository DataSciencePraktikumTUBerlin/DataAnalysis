build plots for the countries we research

```python
countries = ['China', 'Germany', 'India', 'United States']

# Set figsize and loop over our countries
fig, ax = plt.pyplot.subplots(1,1, figsize=(14,6))
for country in countries:
    x = df[df['Country'] == country].Year
    y = df[df['Country'] == country]['Electricity Production (TWh)']
    ax.plot(x, y)

# Set labels and legend
ax.set_title('Generated Electricity Total', size = 20)
ax.legend(countries)
ax.xaxis.set_label_text('Years', size = 16)
ax.yaxis.set_label_text('GigaWatts', size = 16)

# Save the figure. must be saved before .show()
#plt.pyplot.savefig('electricity_production_total.png')

# Show figure
plt.pyplot.show()
```
