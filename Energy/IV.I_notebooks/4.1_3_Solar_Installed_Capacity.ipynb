```python
df = pd.read_csv('c_icaps_NC.csv', index_col=0)
```

```python
df = df[df['Years'] > 2003]
```

```python
df = df[df['Years'] < 2017]
```

build plots for the countries we research

```python
countries = ['China', 'Germany', 'India', 'United States']

# Set figsize and loop over our countries
fig, ax = plt.pyplot.subplots(1,1, figsize=(14,6))
for country in countries:
    x = df[df['Country'] == country].Years
    y = df[df['Country'] == country]['Renewable Energy Solar (Installed capacity) MW']
    ax.plot(x, y)

# Set labels and legend
ax.set_title('Solar Energy installed Capacity', size = 20)
ax.legend(countries)
ax.xaxis.set_label_text('Years', size = 16)
ax.yaxis.set_label_text('MegaWatts', size = 16)

ax.set_xticks(np.arange(min(x), max(x)+1, 2))
# Save the figure in 'IV.I_notebooks' directory. must be saved before .show()
plt.pyplot.savefig('../IV.I_notebooks/3_Solar_Installed_Capacity.png')

# Show figure
plt.pyplot.show()
```
