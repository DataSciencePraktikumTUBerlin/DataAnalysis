build plots for the countries we research

```python
dfg = df.groupby('Year').mean()
dfg = dfg.reset_index()
```

```python

# create figure and axis objects with subplots()
fig,ax = plt.pyplot.subplots(1,1, figsize=(14,6))
ax.grid(color='lightgrey')

# make a plot
ax.plot(dfg['Year'], dfg['Energy intensity level of primary energy (MJ/2011 USD PPP) '], color="red")
# set x-axis label
ax.set_xlabel("year",fontsize=14)
# set y-axis label
ax.set_ylabel("Energy intensity (MJ/USD)",color="red",fontsize=14)

# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(dfg['Year'], dfg['Wind and solar share electricity production (%)'],color="blue")
ax2.set_ylabel("Wind and solar share (%)",color="blue",fontsize=14)
ax.set_title('Energy Intensity vs Wind and Solar share of Electricity Production', size = 20)

plt.pyplot.show()

# save the plot as a file
fig.savefig('IV.I_notebooks/5_Energy_Intensity_vs_Wind_and_Solar_share_Electricity_Production.png', format='jpeg', dpi=100, bbox_inches='tight')
```
