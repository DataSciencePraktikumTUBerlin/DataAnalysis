```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
import seaborn as sns

sns.set_theme(style='darkgrid')
```

```python
df = pd.read_csv('./energy.csv', index_col=0)
df = df[df['Year'] >= 2004]
df = df.groupby('Year').mean()
df
```

```python
def plot(i1, i2):
    y1 = df[i1]
    y2 = df[i2]
    
    i1 = i1.split('(')[0]
    i2 = i2.split('(')[0]
    
    c1 = sns.color_palette()[0]
    c2 = sns.color_palette()[1]

    fig, ax1 = plt.subplots(figsize=(16,8))
    
    ax1 = sns.lineplot(x=df.index.values, y=y1, color=c1)
    ax1.set_ylabel(i1, fontsize=16, color=c1)
    
    ax2 = ax1.twinx()

    ax2 = sns.lineplot(x=df.index.values, y=y2, color=c2)
    ax2.set_ylabel(i2, fontsize=16, color=c2)

    ax2.set_title(f"{i1} & {i2} (œÅ={np.round(y1.corr(y2), 3)})", fontsize=16)
    ax2.set_xlabel('Year')
    ax2.legend()
    
    ax2.grid(False)
```

```python
df
```

```python
i1 = 'Wind and solar share electricity production (%)'
i2 = 'Energy intensity level of primary energy (MJ/2011 USD PPP) '

plot(i1, i2)
```

```python
i1 = 'Total electricity output (GWh) '
i2 = 'Total final energy consumption (TFEC) (TJ) '

plot(i1, i2)
```
