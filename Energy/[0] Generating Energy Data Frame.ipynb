```python
import os

import numpy as np
import pandas as pd
```

```python
folder = './Data Formated'
files = os.listdir(folder)
```

```python
df = pd.DataFrame()

for file in files:
    file = f'{folder}/{file}'
    
    if df.size == 0:
        df = pd.read_csv(file, index_col=0)

        continue

    df2 = pd.read_csv(file, index_col=0)
    df = df.merge(df2, on=['Country', 'Year'])

df = df.replace('..', np.nan).fillna(method='ffill')

df
```

```python
# check if cumilated data

for c in df.columns:
    if c in ['Country', 'Year']:
        continue
        
    print(df[c].is_monotonic_increasing)
```

```python
df.to_csv('./energy.csv')
```
