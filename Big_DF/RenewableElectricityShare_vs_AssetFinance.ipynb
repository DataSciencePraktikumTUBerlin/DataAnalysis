load data

```python
# source_name = 'qu_solar_PV_Power00-19.csv'
d1 = pd.read_csv('result_df/investment_trends.csv', index_col=0) # Asset Finance
d1['Years'] = [2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019] 
d1 = d1[:-4]

d2 = pd.read_csv('result_df/prevc_icaps_NC.csv', index_col=0) # Renewable Energy Solar (Installed capacity) MW
df = pd.merge(d1, d2)
```

build plots for the countries we research

```python
df.columns
```

```python
dfg = df.groupby('Years').mean()
dfg = dfg.reset_index()
```

```python
years = dfg['Years'].array
years
```

```python

# create figure and axis objects with subplots()
fig,ax = plt.pyplot.subplots(1,1, figsize=(14,6))
ax.grid(color='lightgrey')

# make a plot
ax.plot(dfg['Years'], dfg['Renewable Energy Solar (Installed capacity) MW'], color="red", marker="o")
# set x-axis label
ax.set_xlabel("year",fontsize=14)
# set y-axis label
ax.set_ylabel("Installed Solar capacity MW",color="red",fontsize=14)

# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(dfg['Years'], dfg['Asset Finance minus Re-invested equity'],color="blue",marker="o")
ax2.set_ylabel("Asset Finance minus Re-invested equity",color="blue",fontsize=14)
ax.set_title('Renewable Electricity Share vs Project Acquisition & Refinancing', size = 20)

plt.pyplot.show()

# save the plot as a file
fig.savefig('Installed Solar capacity MW vs Project Acquisition & Refinancing.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
```
