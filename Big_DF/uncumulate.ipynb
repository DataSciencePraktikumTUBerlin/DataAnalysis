## Take cumulative data and turn it into non-cumulative data

```python
# Imports
import pandas as pd
import numpy as np
```

```python
pwd
```

```python
# Call desired data
df_meta = pd.read_csv('Indicators_metadata.csv')
source_name = 'prevc_icapw'
df = pd.read_csv('old_result_df/' + source_name + '.csv', index_col=0)  # index_col=0 reads without 'Unnamed: 0'
df
```

```python
# call metadata of the indicator
# df_meta
list(df.columns)
```

```python
# get the indicator column
minus = df['Renewable Energy Wind  (Installed capacity) MW']

# append 4 zeros(number of countries) 
s1 = pd.Series([0,0,0,0])
minus = s1.append(minus)

# create numpy array out of series
x = minus.array
x
```

```python
# delete last 4 entries(number of countries)
y = x[:-4]

# build series from np.array
s = pd.Series(y)
s
```

```python
# add series to df
df['minus'] = s
df
```

```python
# subtract the two rows from another, and by this make the cumulative data non-cumulative
df['Renewable Energy Wind  (Installed capacity) MW'] = df['Renewable Energy Wind  (Installed capacity) MW'] - df['minus']
df
```

```python
# delete the column to subtract
del df['minus']
df
```

```python

# save the updated df. '_NC' postfix stands for non-cumulative
df.to_csv('result_df/'+ source_name +'_NC.csv')
```
