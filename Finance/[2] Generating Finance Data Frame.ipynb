```python
import pandas as pd
```

```python
folder = './Data Formated'
files = ['investment_trends.csv',
        'policies.csv']
```

```python
df = pd.DataFrame()

for file in files:
    file = f'{folder}/{file}'
    
    if df.size == 0:
        df = pd.read_csv(file, index_col=0)

        continue

    df2 = pd.read_csv(file, index_col=0)
    df = df.merge(df2, on=['Country', 'Year'])
    
df
```

```python
df.to_csv('./finance.csv')
```
